# Price vs Neighborhood Group

```python
neighbourhood_mapping = {
    'Bronx': 1,
    'Brooklyn': 2,
    'Manhattan': 3,
    'Queens': 4,
    'Staten Island': 5
}
airbnb['neighbourhood_numeric'] = airbnb['neighbourhood_group'].map(neighbourhood_mapping)
```

* Neighborhood Group = borough
* This dictionary assigns a numerical value to each neighbourhood_group name which is necessary to plot categorical values on the x-axis in a scatter plot.
* Map the neighborhood groups to numerical values

```python
plt.figure(figsize=(12, 6))
plt.scatter(airbnb['neighbourhood_numeric'], airbnb['price'], alpha=0.2) #Alpha - transparent/ opacity
plt.title("Price vs Neighbourhood Group")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Price")
plt.ylim(0, 9000)  # Set a limit

# Customize x-ticks to show neighbourhood names
plt.xticks(ticks=list(neighbourhood_mapping.values()), labels=list(neighbourhood_mapping.keys()))
plt.show()
```

# Price vs Neighborhood 

```python
neighbourhoods = airbnb['neighbourhood'].unique()
neighbourhood_mapping = {neigh: idx for idx, neigh in enumerate(neighbourhoods)}
airbnb['neighbourhood_numeric'] = airbnb['neighbourhood'].map(neighbourhood_mapping)

# Price vs Neighbourhood
plt.figure(figsize=(15, 6))
plt.scatter(airbnb['neighbourhood_numeric'], airbnb['price'], alpha=0.3)
plt.title("Price vs Neighbourhood")
plt.xlabel("Neighbourhood")
plt.ylabel("Price")
plt.ylim(0, 1000)  # Adjust y-axis limit to keep plot readable

# Set x-ticks to display neighbourhood names
plt.xticks(ticks=list(neighbourhood_mapping.values()), labels=list(neighbourhood_mapping.keys()), rotation=90)
plt.show()
```

* creates a dictionary called neighbourhood_mapping that maps each unique neighborhood name to a unique integer index. Hereâ€™s how it works:
* enumerate(neighbourhoods) assigns an index to each neighborhood in the neighbourhoods list.
* {neigh: idx for idx, neigh in ...} creates a dictionary where the neighborhood name (neigh) is the key, and the index (idx) is the value.

```python
plt.figure(figsize=(25, 6))  # Increase width 
plt.scatter(airbnb['neighbourhood_numeric'], airbnb['price'], alpha=0.3)
plt.title("Price vs Neighbourhood")
plt.xlabel("Neighbourhood")
plt.ylabel("Price")
plt.ylim(0, 5000)  # Adjust y-axis limit 

# Set x-ticks to display every 5th neighbourhood label
step = 5  # Show every 5th label to reduce clutter
plt.xticks(ticks=list(neighbourhood_mapping.values())[::step], 
           labels=list(neighbourhood_mapping.keys())[::step], 
           rotation=45, ha='right')

plt.show()
```

* alpha=0.3, Sets transparency to show dense areas more clearly.
* jitter=True, Adds a horizontal spread to avoid overlapping points.
* Adjusts the x-axis ticks to reduce label clutter by showing only every 5th neighborhood label.

# Price per night vs Room Type

```python
# Calculate the average price for each room type
room_type_avg_price = airbnb.groupby('room_type')['price'].mean()

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(room_type_avg_price.index, room_type_avg_price, color=['blue', 'orange', 'green'])
plt.title("Average Price by Room Type")
plt.xlabel("Room Type")
plt.ylabel("Average Price")
plt.ylim(0, 400)  # Set y-limit 

# Add data labels on top of each bar
for i, price in enumerate(room_type_avg_price):
    plt.text(i, price + 20, f"{price:.2f}", ha='center')

plt.show()
```

* This loop iterates over each bar and adds a label above it to display the exact average price
* i is the index of the bar (position on the x-axis).
* price + 20 positions the label slightly above the bar.
* f"{price:.2f}" formats the price to two decimal places.

# Price Distribution Histogram based on their availability

```python
# Separate listings by availability: low, medium, high
low_availability = airbnb[airbnb['availability_365'] <= 100]
medium_availability = airbnb[(airbnb['availability_365'] > 100) & (airbnb['availability_365'] <= 250)]
high_availability = airbnb[airbnb['availability_365'] > 250]

# Plot histograms for each category
plt.figure(figsize=(12, 6))
plt.hist(low_availability['price'], bins=50, alpha=0.5, label='Low Availability (<=100 days)', color='blue')
plt.hist(medium_availability['price'], bins=50, alpha=0.5, label='Medium Availability (101-250 days)', color='orange')
plt.hist(high_availability['price'], bins=50, alpha=0.5, label='High Availability (>250 days)', color='red')

# Add labels and title
plt.title("Price Distribution by Availability")
plt.xlabel("Price per Night")
plt.ylabel("Frequency")
plt.legend(loc='upper right')
plt.xlim(0, 1000)  # Adjust x-axis limit 

plt.show()
```

```python
plt.hist(high_availability['price'], bins=50, alpha=0.5, label='High Availability (>250 days)', color='red')
# Add labels and title
plt.title("Price Distribution by Availability")
plt.xlabel("Price per Night")
plt.ylabel("Frequency")
plt.legend(loc='upper right')
plt.xlim(0, 1000)  # Adjust x-axis limit 
plt.ylim(0, 25000)  
plt.show()
```

# Average Availability vs Neighborhood

```python
# average availability for each neighborhood
neighbourhood_avg_availability = airbnb.groupby('neighbourhood')['availability_365'].mean()

# Plot the bar chart
plt.figure(figsize=(20, 6))  # Increase figure width
plt.bar(neighbourhood_avg_availability.index, neighbourhood_avg_availability, color='teal')
plt.title("Average Availability by Neighborhood")
plt.xlabel("Neighborhood")
plt.ylabel("Average Availability (Days)")

# Show every 5th neighborhood label
step = 5
plt.xticks(ticks=range(0, len(neighbourhood_avg_availability), step),
           labels=neighbourhood_avg_availability.index[::step],
           rotation=45, ha='right')

plt.show()
```

* ticks=range(0, len(neighbourhood_avg_availability), step) sets the tick positions at intervals of 5.
* labels=neighbourhood_avg_availability.index[::step] selects every 5th label from the neighborhood names to be displayed.
